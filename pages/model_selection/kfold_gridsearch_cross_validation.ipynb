Avoid Overfitting

Train data ezberlenmesini önlemek için kullanılır.

```python
from sklearn.datasets import load_iris # iris çiçek datası
from sklearn.model_selection import train_test_split
import pandas as pd 
import numpy as np
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
# normalization
x = (x - np.min(x))/(np.max(x)-np.min(x))
```

```python
# train test split 
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3,random_state=42)
```

```python
# knn model 
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3) # n_neighbours --> k
```

```python
# K fold Cross Validation k=10
from sklearn.model_selection import cross_val_score

# estimators --> (Hangi model kullanılacağını belirtir.) cross_val_score ile çalışırken knn algoitmasınıı kullan.
accuracies = cross_val_score(estimator = knn, X = x_train, y = y_train, cv =10)
```

```python
accuracies
```

```python
print("avarage accuracy: ", np.mean(accuracies))

# yayılım --> ne kadar tutarlı olduğu
print("avarage std: ", np.std(accuracies))
```

Bulunan accuracy ve yayılım değeri ile k değerinin uygunluğuna karar verilir.

```python
# test 
knn.fit(x_train,y_train)
print("test accuracy: ",knn.score(x_test,y_test))
```

Genelde literatürde k değeri 10 seçilir.

# GRID SEARCH CROSS VALIDATION

Yukarıda kullanılan n_neighbours(komşu) sayısı doğru sonucu mu veriyor 3,5 veya başka bir sayı olamaz mı? Bunun anlaşılması için grid search kullanlır.

Birden fazla k değeri ile model tutarlılığı ölçülmesi gerekir.Bundan dolayı GRID SEARCH CROSS VALIDATION ile hem knn algoritmasındaki k değerleri farklı değerler için denenir hem de herbir farklı değer için knn algoritmasını uygularken cross validation yapılır.

<img src="../../img/grid_search_validation.png" width="600" height="270">  

2,3,4,5 değerleri için üçer farklı model oluşturuldu ve 12 farklı grid oluştu buna grid search validation denir.

```python
from sklearn.model_selection import GridSearchCV
```

```python
# n_neighbours(komşu) 1'den 50'ye kadar değerler alsın.
grid = {"n_neighbors":np.arange(1,50)}
```

```python
knn = KNeighborsClassifier()

knn_cv = GridSearchCV(knn, grid, cv=10)
knn_cv.fit(x,y)
```

```python
# print hyperparameter KNN algoritmasındaki K değeri

# tuned --> ayarlanarak bulunmuş
print("tuned hyperparameter K: ",knn_cv.best_params_)
print("tuned parametreye göre en iyi accuracy değeri(best score):,knn,_cv.best_score_ ")
```

# GRID SEARCH CV with LOGISTIC REGRESSION

```python
x = x[:100,:]
y = y[:100]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# l1 --> lasso  l2 --> ridge
grid = {"C":np.logspace(-3,3,7),"penalty":["l1","l2"]} 
```

C çok büyük olduğunda ---> overfit olur

C çok küçük olduğunda ---> underfit

```python
logreg = LogisticRegression()
logreg_cv = GridSearchCV(logreg,grid,cv = 10)
logreg_cv.fit(x,y)
```

```python
print("tuned hyperparameters: (best parameters): ",logreg_cv.best_params_)
print("accuracy:",logreg_cv.best_score_)
```
