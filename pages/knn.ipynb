```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("../data/data.csv")
data
```

M ---> kötü huylu tümör 

B ---> iyi huylu tümör

```python
data.tail()
```

```python
data.drop(["id","Unnamed: 32"], axis=1, inplace=True)
```

```python
M = data[data.diagnosis == "M"]
B = data[data.diagnosis == "B"]
```

```python
# scatter plot 
plt.scatter(M.radius_mean,M.area_mean,color="red",label="kotu")
plt.scatter(B.radius_mean,B.area_mean,color="green",label="iyi")
plt.legend() # label gösterir.
plt.show()
```

```python
# scatter plot 
plt.scatter(M.radius_mean,M.texture_mean,color="red",label="kotu")
plt.scatter(B.radius_mean,B.texture_mean,color="green",label="iyi")
plt.xlabel("radius_mean") # tümör yarıçapı
plt.ylabel("texture_mean") # dokusu
plt.legend() # label gösterir.
plt.show()
```

```python
data.diagnosis = [1 if each == "M" else 0 for each in data.diagnosis]
y = data.diagnosis.values
x_data = data.drop(["diagnosis"],axis=1)
```

```python
# normalization
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data))
```

```python
# train test split 
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.33, random_state=42)
```

```python
# knn modeli 
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn = KNeighborsClassifier(n_neighbors = 3) # n_neighbors = k
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
prediction
```

```python
print(" {} nn score: {} ".format(3,knn.score(x_test,y_test)))
```

%95 accurasy ile doğru tahminler yapıldı.
