```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("../../data/apple.csv")
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.Quality = [1 if each =="good" else 0 for each in df.Quality]
```

```python
df.drop(["A_id"],axis=1,inplace=True)
```

```python
df['Size'] = df['Size'].astype(str).str.replace(".","")

df['Weight'] = df['Weight'].astype(str).str.replace(".","")

df['Sweetness'] = df['Sweetness'].astype(str).str.replace(".","")

df['Crunchiness'] = df['Crunchiness'].astype(str).str.replace(".","")

df['Juiciness'] = df['Juiciness'].astype(str).str.replace(".","")

df['Ripeness'] = df['Ripeness'].astype(str).str.replace(".","")

df['Acidity'] = df['Acidity'].astype(str).str.replace(".","")
```

```python
df['Size'] = df['Size'].astype(float)
df['Size'] = df['Size'].astype('int64')

df['Weight'] = df['Weight'].astype(float)
df['Weight'] = df['Weight'].astype('int64')

df['Sweetness'] = df['Sweetness'].astype(float)
df['Sweetness'] = df['Sweetness'].astype('int64')

df['Crunchiness'] = df['Crunchiness'].astype(float)
df['Crunchiness'] = df['Crunchiness'].astype('int64')

df['Juiciness'] = df['Juiciness'].astype(float)
df['Juiciness'] = df['Juiciness'].astype('int64')

df['Ripeness'] = df['Ripeness'].astype(float)
df['Ripeness'] = df['Ripeness'].astype('int64')

df['Acidity'] = df['Acidity'].astype(float)
df['Acidity'] = df['Acidity'].astype('int64')
```

```python
df
```

```python
y = df.Quality.values
x_data = df.drop(["Quality"],axis=1)
```

```python
x_data
```

```python
x =(x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data))
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
# DoÄŸruluk skoru
print("Decisiontree score: " ,dt.score(x_test,y_test))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 100, random_state=1)
```

n_estimators--> decision tree sayÄ±sÄ±

```python
rf.fit(x_train,y_train)
```

```python
print("random forest algo result: ", rf.score(x_test,y_test))
```
